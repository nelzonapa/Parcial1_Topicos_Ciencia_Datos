```python
import pandas as pd

# Cargar el archivo
file_path = '/content/drive/MyDrive/5TO_ANÃÉO_1ER_SEMESTRE/ToÃÅpicos en Ciencia de Datos/Proyecto - Base de Datos/Activity Logs/ParticipantStatusLogs5.csv'
df = pd.read_csv(file_path)

# Mostrar tipos de datos de todas las columnas
print("Tipos de datos por columna:")
print(df.dtypes)
```

* convertimos la columna 'apartmentId' a tipo entero,
    * apartmentId es un identificador discreto y nominal (nunca deber√≠a tener decimales).
    * Si hubiera valores faltantes, pandas los marca como <NA> en el tipo 'Int64'.
    * Esto nos permitir√° conservar las filas con NA sin perder informaci√≥n, en caso de que falte apartmentId.

```python
import pandas as pd
import glob
import os

base_path = '/content/drive/MyDrive/5TO_ANÃÉO_1ER_SEMESTRE/ToÃÅpicos en Ciencia de Datos/Proyecto - Base de Datos/Activity Logs'
pattern = os.path.join(base_path, 'ParticipantStatusLogs*.csv')

all_logs = []
for filepath in glob.glob(pattern):
    df = pd.read_csv(filepath, encoding='ascii')
    df['apartmentId'] = df['apartmentId'].astype('Int64')

    all_logs.append(df)
```

* Concateno todos los DataFrames en uno solo


```python
logs = pd.concat(all_logs, ignore_index=True)
```

- Cada fila de 'logs' representa un estado de un participante en un tiempo dado.
- apartmentId es una variable discreta y nominal: cada n√∫mero identifica de forma √∫nica un apartamento.
- Queremos saber en qu√© apartmentId pasa m√°s tiempo (en n√∫mero de registros) cada participante.


Pasos:

1. Excluimos cualquier fila donde apartmentId sea NA, pues no contribuye a la asignaci√≥n de "apartamento principal".(Si un registro carece de apartmentId, no sumamos nada; pero mantenemos la fila en 'logs' para cualquier uso futuro).
2. Agrupamos por (participantId, apartmentId) y contamos registros.
3. Para cada participantId, determinamos qu√© apartmentId tiene el conteo m√°s alto.


Justificaci√≥n de tipo de variable:

- apartmentId: discreta (entero), nominal (identificador). No tiene orden inherente.
- participantId: discreta, nominal (identificador).
- El conteo de filas es una variable cuantitativa discreta (1, 2, 3, ‚Ä¶).


Manejamos empates (si un participante tiene el mismo n√∫mero de registros en dos o m√°s apartamentos) eligiendo el apartmentId de menor valor num√©rico arbitrariamente. Esto simplifica la decisi√≥n pero puede cambiarse si hubiera otro criterio (por ejemplo, √∫ltimo registro cronol√≥gico).


NOTA: No eliminamos columnas ni filas originales. Creamos estructuras auxiliares para el conteo.


* Filtrar filas con apartmentId no nulo para el conteo

```python
logs_valid_apartment = logs[logs['apartmentId'].notna()].copy()
```

* Agrupo y cuento apariciones
  - agrupar sobre ['participantId', 'apartmentId'], luego .size() da el conteo

```python
count_por_participante_apto = (
    logs_valid_apartment
    .groupby(['participantId', 'apartmentId'])
    .size()
    .reset_index(name='conteo')
)
```

* Para cada participante, obtener el apartmentId con conteo m√°ximo
  - Ordenamos por participantId asc, conteo desc, apartmentId asc (para desempate).


```python
count_por_participante_apto = count_por_participante_apto.sort_values(
    ['participantId', 'conteo', 'apartmentId'],
    ascending=[True, False, True]
)
```

* Tomamos la primera fila de cada grupo de participantId ‚Üí ese es el apartmentId m√°s frecuente


```python
apartamento_principal = (
    count_por_participante_apto
    .drop_duplicates(subset=['participantId'], keep='first')
    .loc[:, ['participantId', 'apartmentId']]
    .rename(columns={'apartmentId': 'apartmentId_principal'})
    .reset_index(drop=True)
)
```

```python
print(apartamento_principal.head(50))
print(f"Total de participantes con un 'apartmentId_principal' asignado: {len(apartamento_principal)}")
```

Anteriormente ten√≠amos una tabla filtrada por participantes que tienen ni√±os, y con esta √∫ltima lista de participantes con su apartamentId, deberiamos tener en si una tabla de particpantes con su apartemtnID y que tengan ni√±os.

```python
# Unir familias con apartamento_principal
familias_con_apto = familias.merge(
    apartamento_principal,
    on='participantId',
    how='left'  # Mantener todas las filas de familias aunque apartmentId sea desconocido
)

# Mostrar una muestra del resultado
print(familias_con_apto.head())

# Cu√°ntos participantes con ni√±os tienen apartmentId_principal asignado
num_familias_con_apto = familias_con_apto['apartmentId_principal'].notna().sum()
print(f"Participantes con ni√±os que tienen apartmentId principal identificado: {num_familias_con_apto}")
print(f"Total de participantes con ni√±os: {familias_con_apto.shape[0]}")
```

Bien ahora con esa tabla se tiene a los participantes con ni√±os y con sus respectivos apartamentosId , Pero ahora necesitamos saber, quienes son familia. Los mismos participantes que vivan en el mismo apartamento los considerar√© 1 familia.

```python
# Filtrar participantes con apartmentId_principal v√°lido
familias_definidas = familias_con_apto.dropna(subset=['apartmentId_principal']).copy()

# Agrupar por apartmentId_principal para formar familias
familias_definidas['apartmentId_principal'] = familias_definidas['apartmentId_principal'].astype(int)
familias_definidas['familiaId'] = familias_definidas['apartmentId_principal']  # Consideramos cada apto como una familia

# Mostrar una muestra
print(familias_definidas[['participantId', 'apartmentId_principal', 'familiaId']].head())

# ¬øCu√°ntas familias √∫nicas hay?
num_familias = familias_definidas['familiaId'].nunique()
print(f"N√∫mero total de familias (apartamentos √∫nicos con al menos un participante con hijos): {num_familias}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#archivo Apartments.csv (encoding ASCII)
apartments_path = '/content/drive/MyDrive/5TO_ANÃÉO_1ER_SEMESTRE/ToÃÅpicos en Ciencia de Datos/Proyecto - Base de Datos/Attributes/Apartments.csv'
apartments = pd.read_csv(apartments_path, encoding='ascii')
```

* "location" es una variable cualitativa nominal codificada en una estructura tipo texto (cadena de caracteres), con contenido que representa coordenadas. (ASI LO PUSIERON)

* Su forma original "POINT (x y)" no puede ser procesada num√©ricamente ni graficada sin transformaci√≥n. ADEM√ÅS la ubicaci√≥n en el espacio geogr√°fico tiene significado m√©trico: permite comparar distancias, calcular cercan√≠as, concentraciones, proximidades, etc.

Asi que nos conviene transformar a dos variables cuantitativas continuas: x y y.


```python
def parse_point(point_str):
    # elimino las palabras "POINT (" y el otro parentesis ")" para extraer "x y"
    coords = point_str.replace("POINT (", "").replace(")", "").split()
    return float(coords[0]), float(coords[1])
```

Aplico la funci√≥n para extraer 'x' y 'y'


```python
apartments[['x', 'y']] = apartments['location'].apply(lambda loc: pd.Series(parse_point(loc)))
```

2. Ya los grupos separados anteriormente, ahora extraemos su nivel econ√≥mico o sus ganancias.



En si estado financiero, que en si tiene la variable "financialStatus" de ParticipantStatusLogs.
- Pero entonces necesitaremos primero trabajar con las Tablas ParticipantStatusLogs*, verificar si todos los usuarios siempre mantienen su mismo financialStatus.

```python
import glob

base_path = '/content/drive/MyDrive/5TO_ANÃÉO_1ER_SEMESTRE/ToÃÅpicos en Ciencia de Datos/Proyecto - Base de Datos/Activity Logs/'
file_list = glob.glob(base_path + 'ParticipantStatusLogs*.csv')

df_list = []
for fp in file_list:
    df_temp = pd.read_csv(fp, encoding='ascii')
    df_list.append(df_temp)

df_status = pd.concat(df_list, ignore_index=True)

#no duplicados
print("Dimensiones total:", df_status.shape)
print(df_status.dtypes)
```

Verifico valores √∫nicos y faltantes de financialStatus

```python
valores_financ = df_status['financialStatus'].unique()
print("Categor√≠as encontradas en 'financialStatus':", valores_financ)

faltantes_financ = df_status['financialStatus'].isna().sum()
print("Cantidad de valores faltantes en 'financialStatus':", faltantes_financ)
```

Como se tiene un valor con nan, es l√≥gico que pongamos cpmo Unknown

creamos una nueva columna que impute ese NaN a 'Unknown' y la convertimos en category de paso.


```python
df_status['financialStatus_clean'] = df_status['financialStatus'].fillna('Unknown')
df_status['financialStatus_clean'] = df_status['financialStatus_clean'].astype('category')
```

Verifico que ya no hay nan

```python
faltantes_clean = df_status['financialStatus_clean'].isna().sum()
print("NaN en 'financialStatus_clean':", faltantes_clean)
```

Ahora vemos si los participantes mantienen el mismo financialStatus

```python
status_por_participante_clean = (
    df_status
    .groupby('participantId')['financialStatus_clean']
    .nunique()
    .reset_index(name='num_status_unicos_clean')
)

#participantes con m√°s de un estado
cambian_estado_clean = status_por_participante_clean[
    status_por_participante_clean['num_status_unicos_clean'] > 1
]

total_participantes = status_por_participante_clean.shape[0]
num_cambian = cambian_estado_clean.shape[0]
num_mantienen = total_participantes - num_cambian

print(f"De {total_participantes} participantes distintos:")
print(f"- {num_mantienen} mantienen siempre el mismo financialStatus (columna limpia)")
print(f"- {num_cambian} cambian de financialStatus en alg√∫n registro (columna limpia)")
```

- Ordenar por 'participantId' y 'timestamp', y quedarnos con el √∫ltimo registro de cada participante

```python
from google.colab import drive
import os
import pandas as pd

# 1. Montar Google Drive (ejecutar antes que todo)
drive.mount('/content/drive')

# 2. Definir rutas (ajusta si tu carpeta se llama distinto)
base_path = '/content/drive/MyDrive/5TO_ANÃÉO_1ER_SEMESTRE/ToÃÅpicos en Ciencia de Datos/Proyecto - Base de Datos/Attributes/'
restaurants_path = os.path.join(base_path, 'Restaurants.csv')
schools_path     = os.path.join(base_path, 'Schools.csv')
pubs_path        = os.path.join(base_path, 'Pubs.csv')

# 3. Verificar que existan
for p in [restaurants_path, schools_path, pubs_path]:
    if not os.path.isfile(p):
        raise FileNotFoundError(f"No existe el archivo: {p}")

# 4. Cargar los datos
restaurants = pd.read_csv(restaurants_path, encoding='ascii')
schools     = pd.read_csv(schools_path, encoding='ascii')
pubs        = pd.read_csv(pubs_path, encoding='ascii')

# 5. Extraer los valores m√°ximos de costo
max_food_cost    = restaurants['foodCost'].max()     # Costo m√°ximo por comida (USD)
max_monthly_cost = schools['monthlyCost'].max()      # Costo mensual m√°ximo por estudiante (USD)
max_hourly_cost  = pubs['hourlyCost'].max()          # Costo m√°ximo por hora en pub (USD)

# 6. Supuestos para estimar el gasto mensual de la familia:
#    - Familia de 4 personas: 2 adultos (pagan pub y restaurantes) + 2 ni√±os (asisten a la escuela).
#    - Cada miembro come 1 vez al d√≠a en restaurante: 30 d√≠as al mes.
#    - 2 ni√±os pagan el costo mensual escolar m√°ximo cada uno.
#    - Los 2 adultos van al pub 8 veces al mes, consumiendo 1 hora cada vez.
dias_mes               = 30
miembros_familia       = 4
n_adultos              = 2
n_ninos                = 2
visitas_pub_por_adulto = 8
horas_por_visita_pub   = 1

# 7. C√°lculos de gasto mensual:
# 7.1 Restaurante: max_food_cost (USD) √ó 4 personas √ó 30 d√≠as
gasto_mes_restaurante = max_food_cost * miembros_familia * dias_mes

# 7.2 Escuela: max_monthly_cost (USD) √ó 2 ni√±os
gasto_mes_escuela = max_monthly_cost * n_ninos

# 7.3 Pub: max_hourly_cost (USD) √ó (visitas √ó adultos √ó horas)
gasto_mes_pub = max_hourly_cost * visitas_pub_por_adulto * n_adultos * horas_por_visita_pub

# 7.4 Suma total mensual
gasto_mes_total = gasto_mes_restaurante + gasto_mes_escuela + gasto_mes_pub

# 8. Mostrar resultados
print(f"Costo m√°ximo por comida (foodCost): {max_food_cost:.2f} USD")
print(f"Costo mensual m√°ximo por estudiante (monthlyCost): {max_monthly_cost:.2f} USD")
print(f"Costo m√°ximo por hora en pub (hourlyCost): {max_hourly_cost:.2f} USD\n")

print("Supongo el siguiente caso")
print(f"- Cada persona come 1 vez al d√≠a ‚Üí {miembros_familia} √ó {dias_mes} comidas al mes.")
print(f"- {n_ninos} ni√±os asisten a la escuela y pagan mensualidad.")
print(f"- {n_adultos} adultos van al pub {visitas_pub_por_adulto} veces al mes, 1 hora cada vez.\n")

print("== Gasto Mensual Estimado ==")
print(f"- Restaurante: {gasto_mes_restaurante:.2f} USD")
print(f"- Escuela:    {gasto_mes_escuela:.2f} USD")
print(f"- Pub:        {gasto_mes_pub:.2f} USD")
print(f"‚Üí Total:     {gasto_mes_total:.2f} USD")
```

```python
from google.colab import drive
import pandas as pd
import geopandas as gpd
from shapely import wkt
import matplotlib.pyplot as plt

# 1. Montar Google Drive (ejecutar esta celda primero)
drive.mount('/content/drive')

# 2. Rutas a los archivos
base_path = '/content/drive/MyDrive/5TO_ANÃÉO_1ER_SEMESTRE/ToÃÅpicos en Ciencia de Datos/Proyecto - Base de Datos/Attributes/'
restaurants_path = base_path + 'Restaurants.csv'
schools_path     = base_path + 'Schools.csv'
pubs_path        = base_path + 'Pubs.csv'
buildings_path   = base_path + 'Buildings.csv'

# 3. Cargar los CSVs con encoding ASCII
restaurants = pd.read_csv(restaurants_path, encoding='ascii')
schools     = pd.read_csv(schools_path, encoding='ascii')
pubs        = pd.read_csv(pubs_path, encoding='ascii')
buildings   = pd.read_csv(buildings_path, encoding='ascii')

# 4. Convertir la columna 'location' de WKT a geometr√≠a
# Para restaurantes, escuelas y pubs son puntos
restaurants['geometry'] = restaurants['location'].apply(wkt.loads)
schools['geometry']     = schools['location'].apply(wkt.loads)
pubs['geometry']        = pubs['location'].apply(wkt.loads)

# Para edificios, la geometr√≠a es un POLYGON
buildings['geometry']   = buildings['location'].apply(wkt.loads)

# 5. Crear GeoDataFrames
gdf_restaurants = gpd.GeoDataFrame(restaurants, geometry='geometry')
gdf_schools     = gpd.GeoDataFrame(schools, geometry='geometry')
gdf_pubs        = gpd.GeoDataFrame(pubs, geometry='geometry')
gdf_buildings   = gpd.GeoDataFrame(buildings, geometry='geometry')

# 6. Graficar todas las capas en un mismo mapa
fig, ax = plt.subplots(figsize=(12, 12))

# Edificios (pol√≠gonos) con borde gris y transparencia para no bloquear puntos
gdf_buildings.plot(ax=ax, facecolor='none', edgecolor='gray', linewidth=0.5, label='Buildings')

# Puntos de Restaurantes, Escuelas y Pubs
gdf_restaurants.plot(ax=ax, color='red', marker='o', label='Restaurantes', markersize=50)
gdf_schools.plot(ax=ax, color='blue', marker='s', label='Escuelas', markersize=50)
gdf_pubs.plot(ax=ax, color='green', marker='^', label='Pubs', markersize=50)

# Etiquetas y leyenda
plt.title('Ubicaci√≥n de Restaurantes, Escuelas, Pubs y Edificios')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.legend()

plt.show()
```
